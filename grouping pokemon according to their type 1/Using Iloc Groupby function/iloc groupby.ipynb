```python
import pandas as pd
print('**Grouping of pokemons on the basis of their type 1 using iloc**')
df=pd.read_csv("C:\\Users\\SYED\\Desktop\\Aedifico\\24-10-2019\\CSV Files\\Pokemon.csv")
```
