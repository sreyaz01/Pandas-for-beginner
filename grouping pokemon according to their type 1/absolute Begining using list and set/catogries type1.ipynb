```python
import pandas as pd
print('**Grouping of pokemons on the basis of their type 1**')
df=pd.read_csv("C:\\Users\\SYED\\Desktop\\Aedifico\\24-10-2019\\CSV Files\\Pokemon.csv")

a=df.Type1 #storing column in a variable
Set_convert=set(a) #converting columns into set because set doesn't support duplicacy
print(Set_convert)
```

```python
list_convert=list(Set_convert) #converting set into list
var_list=list.copy(list_convert) # a list for auto assign variables basis on type1
print(var_list)
```

```python
for i in range (len(list_convert)):
    var_list[i]=df.loc[df["Type1"]==list_convert[i]]
    df1=pd.DataFrame(var_list[i])
    print(df1)
    
    #also this code is for ultimate begineer using set and list to sorting a value just for logical purpose there is a method for this also you can see that in another files
    #their is an issue i want to print result by acessing type for now result is like sorting values
```
