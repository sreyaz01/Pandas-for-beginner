<b>Problem Statement : from the given car.CSV file Find Car which has 'Sport' in Their model Name?</b>

```python
import pandas as pd
print('**Finding Cars which Model Name Contains "Sports" in There name**')
```

```python
df=pd.read_csv("C:\\Users\\SYED\\Desktop\\Aedifico\\24-10-2019\\CSV Files\\car.csv")
```

```python
df1=df[df['Model'].str.contains('sport',case=False)] #to find particular substring we use str.contains(), SYNTAX: datafram_name[datframe_name['column'].str.contains('substring')] case=false that means no casesensitive search
```
